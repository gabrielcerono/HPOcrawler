```python
import os
from selenium import webdriver
import pandas as pd
import time
from selenium.webdriver.support.ui import Select
```

```python
DONWLOAD_FOLDER = os.path.join(os.getcwd(), "downloads")
```

```python
chromeOptions = webdriver.ChromeOptions()
prefs = {"download.default_directory": DONWLOAD_FOLDER}
chromeOptions.add_experimental_option("prefs", prefs)
DRIVER = webdriver.Chrome(options=chromeOptions)
```

```python
KEYWORD = "serum"
```

```python
DRIVER.get(f'https://hpo.jax.org/app/browse/search?q={KEYWORD}&navFilter=all')
```

```python
def get_all_rows():
    "Changes the row values from 50, to 500"
    DRIVER.find_element_by_id("mat-select-0").click()
    time.sleep(1)
    DRIVER.find_element_by_id("mat-option-2").click()
```

```python
get_all_rows()
```

```python
headers = ["Term Identifier", "Term Name", "Matching String", "No. Descendants"]
```

```python
def get_pandas_rows(header):
    "Collect the table, and parse it into a pandas data frame"
    rows = DRIVER.find_elements_by_class_name('mat-row.cdk-row.ng-star-inserted')
    serum = pd.DataFrame([row.text.split("\n") for row in rows], columns=headers)
    return serum
```

```python
serum_df = get_pandas_rows(headers)
```

```python
serum_df
```

## Crawler + Parser

```python
def crawler(url, chromeOptions):
    driver = webdriver.Chrome(options=chromeOptions)
    driver.get(url)
    time.sleep(2)
    driver.find_element_by_class_name("mat-focus-indicator.download-associations.mat-raised-button.mat-button-base").click()
    time.sleep(2)
    try:
        driver.find_elements_by_class_name("mat-focus-indicator.dialog-selection.mat-raised-button.mat-button-base")[0].click()
        time.sleep(3)
    except IndexError:
        print('Esta url no tiene dataset')
    driver.quit()
```

```python
for x in serum_df["Term Identifier"]:
    url = "https://hpo.jax.org/app/browse/term/" + x
    crawler(url, chromeOptions)
```

```python
dfs = []
for xlsx in os.listdir('downloads'):
    lab_term = xlsx[13:23].replace("_", ":")
    df = pd.read_excel("downloads/" + xlsx)
    df['Term Identifier'] = lab_term
    dfs.append(df)
    
```

```python
complete_disease_edges = pd.concat(dfs)
```

```python
complete_disease_edges
```

```python
grouped = complete_disease_edges[["DISEASE_ID", "Term Identifier"]].groupby(by=["Term Identifier"]).count()
```

```python
grouped
```

```python
serum_complete = serum_df
```

```python
serum_complete = serum_complete.merge(grouped, how='left', on="Term Identifier")
```
