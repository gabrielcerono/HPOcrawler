```python
import pandas as pd
import numpy as np
```

```python
cross_ref = pd.read_csv('xrefs.tsv', sep='\t')
```

```python
cross_ref.head()
```

Select rows by OMIM CrossRef

```python
OMIM_DO = cross_ref[cross_ref.resource == 'OMIM']
```

```python
OMIM_DO.head()
```

HPO - OMIM File

```python
hpo_crossref = pd.read_csv("phenotype_annotation.tab", sep='\t', low_memory= False)
```

```python
hpo_crossref.head()
```

```python
hpo_omim = hpo_crossref[hpo_crossref['reference'].str.contains("OMIM", na = False)]
```

```python
hpo_omim.rename(columns = {"#disease-db" : "disease_db", "disease-identifier": "disease_identifier"}, inplace = True)
```

```python
hpo_omim.head()
```

```python
hpo_omim_final = hpo_omim[['disease_db', 'disease_identifier', 'HPO-ID']]
```

```python
hpo_omim_final.head()
```

```python
hpo_omim_final.dtypes
```

Uniting the dataframe dataframes

```python
OMIM_DO.rename(columns = {"resource_id" : "disease_identifier"}, inplace = True)
```

```python
hpo_to_do = pd.merge(OMIM_DO.astype({'disease_identifier': 'int64'}), hpo_omim_final, on = 'disease_identifier')
```

```python
hpo_to_do.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='hpo-to-do.csv')  
hpo_to_do.to_csv('hpomerged.zip', index=False,
          compression=compression_opts) 
```
